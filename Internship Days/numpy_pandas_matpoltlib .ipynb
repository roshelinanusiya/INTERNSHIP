NumPy Basic Program

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50])
print("NumPy Array:")
print(arr)
print("Addition:", arr + 5)
print("Multiplication:", arr * 2)
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
```

List vs Array Program

```python
import numpy as np
list1 = [1, 2, 3, 4]
array1 = np.array([1, 2, 3, 4])
print("Python List:")
print(list1)
print("NumPy Array:")
print(array1)
print("List Multiplication:")
print(list1 * 2)
print("Array Multiplication:")
print(array1 * 2)
```

Pandas Basic Program

```python
import pandas as pd
data = {
    "Name": ["Anu", "John", "Rahul"],
    "Age": [20, 21, 22],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Average Marks:", df["Marks"].mean())
```

Matplotlib Line Graph

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.title("Line Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

Matplotlib Bar Graph

```python
import matplotlib.pyplot as plt
students = ["A", "B", "C", "D"]
marks = [80, 90, 70, 85]
plt.bar(students, marks)
plt.title("Bar Graph")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

Matplotlib Pie Chart

```python
import matplotlib.pyplot as plt
subjects = ["Python", "Java", "C++"]
students = [40, 35, 25]
plt.pie(students, labels=subjects, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

Matplotlib Scatter Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4 , 5]
y = [5, 10, 15, 20, 25]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

Histogram Program

```python
import matplotlib.pyplot as plt
marks = [10, 20, 20, 30, 40, 40, 40, 50]
plt.hist(marks)
plt.title("Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

Box Plot Program

```python
import matplotlib.pyplot as plt
data = [10, 20, 30, 40, 50]
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

Simple Subplot Program

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Line")
plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.title("Bar")
plt.show()
```
